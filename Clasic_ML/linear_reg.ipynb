### here we need to calculate cost function 
$f_{w,b}(x^{(i)})$: $$f_{w,b}(x^{(i)}) = wx^{(i)} + b \tag{1}$$
The cost function is:

$[
J(w, b) = \frac{1}{2m} \sum_{i=1}^{m} \left( w \cdot x^{(i)} + b - y^{(i)} \right)^2
]$


```python
def cost_function(w,b,x,y):
    m=x.shape[0]
    cost=0
    for i in range(m):
        f_wb=w*x[i]+b
        cost=cost+(f_wb-y[i])**2
    final_cost=(1/(2*m))*cost
    return final_cost
```


The gradient is defined as:
$$
\begin{align}
\frac{\partial J(w,b)}{\partial w}  &= \frac{1}{m} \sum\limits_{i = 0}^{m-1} (f_{w,b}(x^{(i)}) - y^{(i)})x^{(i)} \tag{4}\\
  \frac{\partial J(w,b)}{\partial b}  &= \frac{1}{m} \sum\limits_{i = 0}^{m-1} (f_{w,b}(x^{(i)}) - y^{(i)}) \tag{5}\\
\end{align}
$$

Here *simultaniously* means that you calculate the partial derivatives for all the parameters before updating any of the parameters.

<a name="toc_40291_2.3"></a>
### compute_gradient
<a name='ex-01'></a>
`compute_gradient`  implements (4) and (5) above and returns $\frac{\partial J(w,b)}{\partial w}$,$\frac{\partial J(w,b)}{\partial b}$.

```python
def compute_gradient(x,y,w,b):
    m=x.shape[0]
    dj_dw=0
    dj_db=0
    for i in range (m):
        f_wb=w*x[i]+b
        dj_dw_i=(f_wb-y[i])*x[i]
        dj_db_i=(f_wb-y[i])
        dj_dw+=dj_dw_i       #& dj_dw_i is for ith example in dataset
        dj_db+=dj_db_i       #& dj_db_i is for ith example in dataset
    dj_dw=dj_dw/m
    dj_db=dj_db/m
    return dj_dw , dj_db
    
        
```

*gradient descent* was described as:

$$\begin{align*} \text{repeat}&\text{ until convergence:} \; \lbrace \newline
\;  w &= w -  \alpha \frac{\partial J(w,b)}{\partial w} \tag{3}  \; \newline 
 b &= b -  \alpha \frac{\partial J(w,b)}{\partial b}  \newline \rbrace
\end{align*}$$
where, parameters $w$, $b$ are updated simultaneously.  

```python
def gradient_descent(w_init,b_init,alpha,x,y,num_iters):        #& num_iters-> num of iterations
    w=w_init
    b=b_init
    #& An array to store cost J and w's at each iteration primarily for graphing later
    for i in range(num_iters):
        j_hist=[]   # cost function history
        p_hist=[]   # pair of parameters [w,b]
        dj_dw,dj_db=compute_gradient(x,y,w,b)
        b=b-alpha*dj_db
        w=w-alpha*dj_dw
        if i<=100000:     # prevent resource exhaustion 
            j_hist.append(cost_function(w,b,x,y))
            p_hist.append([w,b])
        if i%math.ceil(num_iters/10)==0:
            print(f"Iteration {i:4}: Cost {j_hist[-1]:0.2e} ",
                  f"dj_dw: {dj_dw: 0.3e}, dj_db: {dj_db: 0.3e}  ",
                  f"w: {w: 0.3e}, b:{b: 0.5e}")
    return w, b, j_hist, p_hist #return w and J,w history for graphing
    
```

```python
w_init,b_init=0,0
alpha=1.0e-2
iterations=10000
w_final,b_final,j_hist,p_hist = gradient_descent(w_init,b_init,alpha,x_train,y_train,iterations)

print(f"the final w, b found are ({w_final:8.4f},{b_final:8.4f})")
```
